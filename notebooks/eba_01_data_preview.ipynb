```python
import pandas as pd

df = pd.read_csv("../data/raw/2023_hungary_race_laps.csv")

# Convert lap and sector times to timedelta (if not already)
for col in ['LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time', 'PitInTime', 'PitOutTime']:
    if col in df.columns:
        df[col] = pd.to_timedelta(df[col])

# Convert numeric columns
numeric_cols = ['SpeedI1', 'SpeedI2', 'SpeedFL', 'SpeedST', 'LapNumber', 'Stint']
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print(df.info())
print(df.head())
```
